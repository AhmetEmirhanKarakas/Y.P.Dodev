```python
#EĞİTİM
import os
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image

class CustomTestDataset(Dataset):
    def __init__(self, directory, transform=None):
        self.directory = directory
        self.transform = transform
        self.images = []
        self.labels = []

        # Test klasöründeki tüm resimleri yükle
        for filename in os.listdir(directory):
            if filename.endswith('.jpg') or filename.endswith('.png'):  # Resim dosyalarını filtrele
                image_path = os.path.join(directory, filename)
                self.images.append(image_path)

                # Fotoğrafın adından veya başka bir özelliğinden etiket belirleyin
                if 'dirty' in filename.lower():
                    self.labels.append(0)  # dirty için etiket
                elif 'cleaned' in filename.lower():
                    self.labels.append(1)  # cleaned için etiket
                else:
                    self.labels.append(-1)  # bilinmeyen resimler için etiket (eğer varsa)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img = Image.open(self.images[idx])
        label = self.labels[idx]
        if self.transform:
            img = self.transform(img)
        return img, label

# Test veri kümesini yüklemek için:
test_dataset = CustomTestDataset('/content/drive/My Drive/Data/plates/test/', transform=transform)

# DataLoader oluştur
test_loader = DataLoader(test_dataset, batch_size=32)

# Test veri yükleyicisini kontrol et
for data_batch in test_loader:
    print("Batch Shape (Images):", data_batch[0].shape)  # Test verisinin boyutları
    print("Batch Labels:", data_batch[1])               # Test etiketleri
    break
```

```python
import torch
import torch.nn as nn

class CustomModel(nn.Module):
    def __init__(self):
        super(CustomModel, self).__init__()
        # Örnek bir CNN yapısı
        self.conv1 = nn.Conv2d(3, 64, kernel_size=3, stride=1, padding=1)
        self.conv2 = nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1)
        self.fc1 = nn.Linear(128 * 32 * 32, 256)  # 128 * 32 * 32 burada son konvolüsyonel katmandan gelen özellik haritası boyutuna göre ayarlanmalı
        self.fc2 = nn.Linear(256, 10)  # Örnek: 10 sınıf

    def forward(self, x):
        # Konvolüsyonel katmanlar
        x = self.conv1(x)
        x = torch.relu(x)
        x = self.conv2(x)
        x = torch.relu(x)

        # Düzleştirme (Flatten)
        x = x.view(x.size(0), -1)  # (batch_size, -1) şeklinde düzeltilir.

        # Tam bağlantılı katmanlar
        x = self.fc1(x)
        x = torch.relu(x)
        x = self.fc2(x)
        return x
```

```python
# Lightning DataModule
class CustomDataModule(pl.LightningDataModule):
    def __init__(self, root_dir, batch_size=32, transform=transform):
        """
        Args:
            root_dir (str): Dataset'in kök dizini.
            batch_size (int): Mini-batch boyutu.
            transform (torchvision.transforms): Verilere uygulanacak dönüşümler.
        """
        super().__init__()
        self.root_dir = root_dir
        self.batch_size = batch_size
        self.transform = transform

    def setup(self, stage=None):
        """
        Verileri eğitim ve test setlerine ayırır.
        Args:
            stage (str): 'fit' (eğitim) veya 'test' aşaması.
        """
        # Tüm veri setini ImageFolder ile yükle
        full_dataset = datasets.ImageFolder(root=self.root_dir, transform=self.transform)

        # Dataset'i sınıflandırma için ayır
        targets = [sample[1] for sample in full_dataset.samples]  # Sınıf etiketlerini al
        train_indices, test_indices = train_test_split(
            list(range(len(full_dataset))),
            test_size=0.2,
            stratify=targets
        )

        # Subset kullanarak veri setlerini ayır
        self.train_dataset = Subset(full_dataset, train_indices)
        self.test_dataset = Subset(full_dataset, test_indices)

    def train_dataloader(self):
        """Eğitim verisi için DataLoader döndürür."""
        return DataLoader(self.train_dataset, batch_size=self.batch_size, shuffle=True, num_workers=4)

    def test_dataloader(self):
        """Test verisi için DataLoader döndürür."""
        return DataLoader(self.test_dataset, batch_size=self.batch_size, shuffle=False, num_workers=2)
```

```python
# EĞİTİM

# CNN Modeli
class ModifiedCNN(nn.Module):
    def __init__(self):
        super(ModifiedCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 32, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(32, 64, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2),
            nn.Conv2d(64, 128, kernel_size=3, stride=1, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(kernel_size=2, stride=2)
        )
        self.classifier = nn.Sequential(
            nn.Linear(128 * 28 * 28, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, 128),
            nn.ReLU(),
            nn.Linear(128, len(class_names))
        )

    def forward(self, X):
        X = self.features(X)
        X = X.view(X.size(0), -1)
        X = self.classifier(X)
        return X

    def configure_optimizers(self):
        optimizer = torch.optim.Adam(self.parameters(), lr=0.001)
        return optimizer

    def training_step(self, train_batch, batch_idx):
        X, y = train_batch
        y_hat = self(X)
        loss = F.cross_entropy(y_hat, y)
        self.log("train_loss", loss)
        return loss

    def test_step(self, test_batch, batch_idx):
        X, y = test_batch
        y_hat = self(X)
        loss = F.cross_entropy(y_hat, y)
        self.log("test_loss", loss)
```

```python
# EĞİTİM
if __name__ == '__main__':
    # Veri modülünü başlat
    datamodule = DataModule()
    datamodule.setup()

    # Modeli tanımla
    model = ConvolutionalNetwork()

    # Cihaz seçimi (GPU varsa kullan, yoksa CPU kullan)
    device_count = 1 if torch.cuda.is_available() else 1  # GPU için 1 cihaz, CPU için de 1 cihaz

    # Trainer
    trainer = Trainer(
        max_epochs=50,                              # Eğitim için epoch sayısı
        accelerator="gpu" if torch.cuda.is_available() else "cpu",  # GPU veya CPU kullan
        devices=device_count                       # Kullanılacak cihaz sayısı
    )

    # Modeli eğit
    trainer.fit(model, datamodule)

    # Test verisini ayarla ve modeli test et
    datamodule.setup(stage='test')
    test_loader = datamodule.test_dataloader()
    trainer.test(dataloaders=test_loader)
```
